# Data Preprocessing

```python
# 1. Clinical data processing
#   - (Patient ID, Overall patient survival status, Overall survival in months since initial diagnosis)

clinical_df = clinical_df[['PATIENT_ID', 'OS_STATUS', 'OS_MONTHS']]
clinical_df.head()
```

```python
# 2. RSEM Data Processing

# df_expressionì€ ìœ ì „ìž ë°œí˜„ëŸ‰ ë°ì´í„°í”„ë ˆìž„ìž…ë‹ˆë‹¤.
# 'Hugo_Symbol'ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì •í•œ í›„ 'EGR1' í–‰ì„ ì„ íƒí•©ë‹ˆë‹¤.
# ê·¸ í›„ PATIENT_IDë¥¼ ì»¬ëŸ¼ìœ¼ë¡œ ë§Œë“¤ì–´ì¤ë‹ˆë‹¤.
rsem_df = rsem_df.set_index('Hugo_Symbol').loc['EGR1'].to_frame(name='EGR1_EXPRESSION')
rsem_df = rsem_df.reset_index().rename(columns={'index': 'PATIENT_ID'})

rsem_df['PATIENT_ID'] = rsem_df['PATIENT_ID'].str[:12]

rsem_df
```

```python
# Merge two dataframe

# 'inner' ë°©ì‹ìœ¼ë¡œ ë³‘í•©í•˜ì—¬ ë‘ ë°ì´í„°ì— ëª¨ë‘ ì¡´ìž¬í•˜ëŠ” í™˜ìž ì •ë³´ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
final_df = pd.merge(clinical_df, rsem_df, on='PATIENT_ID', how='inner')
```

```python
final_df
```

```python
# ê²°ì¸¡ì¹˜ê°€ ìžˆëŠ” í–‰ì„ ì œê±°í•©ë‹ˆë‹¤.
final_df = final_df.dropna()

# OS_STATUSê°€ 'DECEASED'ì´ë©´ 1, 'LIVING'ì´ë©´ 0ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
final_df['EVENT'] = final_df['OS_STATUS'].str.contains('DECEASED').astype(int)

# ìµœì¢… ë°ì´í„° í™•ì¸
print("ì „ì²˜ë¦¬ ì™„ë£Œëœ ìµœì¢… ë°ì´í„°:")
final_df
```

```python
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
import matplotlib.pyplot as plt

# 1. ê·¸ë£¹ ë‚˜ëˆ„ê¸°
median_expression = final_df['EGR1_EXPRESSION'].median()
high_group_filter = final_df['EGR1_EXPRESSION'] >= median_expression
low_group_filter = final_df['EGR1_EXPRESSION'] < median_expression

# 2. ì¹´í”Œëž€-ë§ˆì´ì–´ ìƒì¡´ ê³¡ì„  ì‹œê°í™”
kmf_high = KaplanMeierFitter()
kmf_low = KaplanMeierFitter()

ax = plt.subplot(111)

# High ê·¸ë£¹
kmf_high.fit(durations=final_df[high_group_filter]['OS_MONTHS'],
             event_observed=final_df[high_group_filter]['EVENT'],
             label=f'EGR1 High (n={sum(high_group_filter)})')
kmf_high.plot_survival_function(ax=ax)

# Low ê·¸ë£¹
kmf_low.fit(durations=final_df[low_group_filter]['OS_MONTHS'],
            event_observed=final_df[low_group_filter]['EVENT'],
            label=f'EGR1 Low (n={sum(low_group_filter)})')
kmf_low.plot_survival_function(ax=ax)

# ê·¸ëž˜í”„ ì œëª© ë° ë¼ë²¨ ì„¤ì •
ax.set_title('AML Patient Survival based on EGR1 Expression')
ax.set_xlabel('Time (Months)')
ax.set_ylabel('Survival Probability')
ax.grid(True)
plt.show()

# 3. ë¡œê·¸-ìˆœìœ„ í†µê³„ ê²€ì •
results = logrank_test(
    durations_A=final_df[high_group_filter]['OS_MONTHS'],
    durations_B=final_df[low_group_filter]['OS_MONTHS'],
    event_observed_A=final_df[high_group_filter]['EVENT'],
    event_observed_B=final_df[low_group_filter]['EVENT']
)

print("\n--- Log-Rank Test Results ---")
results.print_summary()
```

```python
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
import matplotlib.pyplot as plt

# 1. Define the expression thresholds for the top and bottom 20%
low_quantile = final_df['EGR1_EXPRESSION'].quantile(0.2)
high_quantile = final_df['EGR1_EXPRESSION'].quantile(0.8)

# 2. Create filters for the two groups
low_20_filter = final_df['EGR1_EXPRESSION'] <= low_quantile
high_20_filter = final_df['EGR1_EXPRESSION'] >= high_quantile

# 3. Perform Kaplan-Meier survival analysis and visualization ðŸ“Š
kmf = KaplanMeierFitter()
ax = plt.subplot(111)

# Plot for the Highest 20% expression group
kmf.fit(durations=final_df[high_20_filter]['OS_MONTHS'],
        event_observed=final_df[high_20_filter]['EVENT'],
        label=f'EGR1 High 20% (n={sum(high_20_filter)})')
kmf.plot_survival_function(ax=ax)

# Plot for the Lowest 20% expression group
kmf.fit(durations=final_df[low_20_filter]['OS_MONTHS'],
        event_observed=final_df[low_20_filter]['EVENT'],
        label=f'EGR1 Low 20% (n={sum(low_20_filter)})')
kmf.plot_survival_function(ax=ax)

# Set graph titles and labels
ax.set_title('Survival of Top 20% vs. Bottom 20% EGR1 Expression')
ax.set_xlabel('Time (Months)')
ax.set_ylabel('Survival Probability')
ax.grid(True)
plt.show()

# 4. Perform the log-rank test for statistical significance ðŸ“
results = logrank_test(
    durations_A=final_df[high_20_filter]['OS_MONTHS'],
    durations_B=final_df[low_20_filter]['OS_MONTHS'],
    event_observed_A=final_df[high_20_filter]['EVENT'],
    event_observed_B=final_df[low_20_filter]['EVENT']
)

print("\n--- Log-Rank Test Results ---")
results.print_summary()
```
