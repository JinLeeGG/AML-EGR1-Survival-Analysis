# Building a Machine Learning Model that predicts patient's status based on RNA Expression data




```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
merged_df = pd.read_csv('/content/drive/MyDrive/Acute Myeloid Leukemia (TCGA, PanCancer Atlas)/Processed_Data/merged_df.csv')
merged_df.dropna(subset=['Observation Period'], inplace=True)
merged_df
```

```python
# Define the target variable
y = merged_df['Status']

# Define features by dropping the non-gene and target columns
X = merged_df.drop(columns=['bcr_patient_barcode', 'Status', 'Observation Period'])

# Display the shapes to verify
print("Features shape (X):", X.shape)
print("Target shape (y):", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split the data, keeping the proportion of classes the same (stratify)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2025, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Initialize and train the model
log_reg = LogisticRegression(random_state=2025, max_iter=1000)
log_reg.fit(X_train, y_train)

# Make predictions
y_pred_lr = log_reg.predict(X_test)

# Evaluate the model
print("--- Logistic Regression Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.2f}")
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
rand_forest = RandomForestClassifier(random_state=2025, n_estimators=100)
rand_forest.fit(X_train, y_train)

# Make predictions
y_pred_rf = rand_forest.predict(X_test)

# Evaluate the model
print("--- Random Forest Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.2f}")
print(classification_report(y_test, y_pred_rf))
```

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
from lifelines.statistics import logrank_test
import pandas as pd
from tqdm import tqdm # ë£¨í”„ ì§„í–‰ ìƒí™©ì„ ì‹œê°ì ìœ¼ë¡œ ë³´ì—¬ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬

# --- ë°ì´í„° ì¤€ë¹„ ---
# 'merged_df'ëŠ” ìž„ìƒ ì •ë³´ì™€ ìœ ì „ìž ë°œí˜„ëŸ‰ì´ í•©ì³ì§„ ë°ì´í„°í”„ë ˆìž„ìž…ë‹ˆë‹¤.

# 1. íŠ¹ì„±(X) ë°ì´í„° ì¤€ë¹„: merged_dfì—ì„œ ìœ ì „ìž ë°œí˜„ëŸ‰ì— í•´ë‹¹í•˜ì§€ ì•ŠëŠ” ì»¬ëŸ¼ë“¤ì„ ì œì™¸í•©ë‹ˆë‹¤.
X_full = merged_df.drop(columns=['bcr_patient_barcode', 'Status', 'Observation Period'])

# 2. ìƒì¡´ ì •ë³´(ì‹œê°„, ì´ë²¤íŠ¸)ê°€ ë‹´ê¸´ ë°ì´í„°í”„ë ˆìž„ì„ ì§€ì •í•©ë‹ˆë‹¤.
df_survival_final = merged_df

# ê²°ê³¼ë¥¼ ì €ìž¥í•  ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
results = []

# tqdmì„ ì‚¬ìš©í•˜ì—¬ ì „ì²´ ìœ ì „ìžì— ëŒ€í•œ ë°˜ë³µë¬¸ì˜ ì§„í–‰ ìƒí™©ì„ í™•ì¸í•©ë‹ˆë‹¤.
for gene in tqdm(X_full.columns, desc="ëª¨ë“  ìœ ì „ìž ë¶„ì„ ì¤‘ (Log-rank)"):
    try:
        # í˜„ìž¬ ìœ ì „ìžì˜ ë°œí˜„ëŸ‰ ë°ì´í„°
        gene_expression = X_full[gene]

        # ì¤‘ì•™ê°’(median)ì„ ê¸°ì¤€ìœ¼ë¡œ ê³ ë°œí˜„/ì €ë°œí˜„ ê·¸ë£¹ì„ ë‚˜ëˆ•ë‹ˆë‹¤.
        median_expression = gene_expression.median()

        # ë§Œì•½ ì¤‘ì•™ê°’ì´ 0ì´ë©´ (ëŒ€ë¶€ë¶„ì˜ ê°’ì´ 0ì¸ ê²½ìš°), ë¶„ì„ì—ì„œ ì œì™¸í•˜ê³  ë‹¤ìŒ ìœ ì „ìžë¡œ ë„˜ì–´ê°‘ë‹ˆë‹¤.
        if median_expression == 0:
            continue

        # ê·¸ë£¹ í•„í„° ìƒì„±
        high_group_filter = gene_expression >= median_expression
        low_group_filter = gene_expression < median_expression

        # ë¡œê·¸-ìˆœìœ„ ê²€ì • ìˆ˜í–‰ (ì‚¬ìš©ìž ë°ì´í„°ì˜ ì»¬ëŸ¼ëª…ìœ¼ë¡œ ìˆ˜ì •)
        result = logrank_test(
            durations_A=df_survival_final.loc[high_group_filter, 'Observation Period'],
            durations_B=df_survival_final.loc[low_group_filter, 'Observation Period'],
            event_observed_A=df_survival_final.loc[high_group_filter, 'Status'],
            event_observed_B=df_survival_final.loc[low_group_filter, 'Status']
        )

        # ê²°ê³¼(ìœ ì „ìž ì´ë¦„, p-value)ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì €ìž¥í•©ë‹ˆë‹¤.
        results.append({'gene': gene, 'p_value': result.p_value})

    except Exception as e:
        # ë¶„ì„ ì¤‘ ì—ëŸ¬ê°€ ë°œìƒí•˜ë”ë¼ë„ ì „ì²´ ê³¼ì •ì´ ë©ˆì¶”ì§€ ì•Šë„ë¡ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        # print(f"Error analyzing {gene}: {e}") # ì—ëŸ¬ ì›ì¸ì„ í™•ì¸í•˜ê³  ì‹¶ì„ ë•Œ ì£¼ì„ í•´ì œ
        continue

# ê²°ê³¼ë¥¼ ë°ì´í„°í”„ë ˆìž„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
results_df = pd.DataFrame(results)

# p-valueê°€ 0.05 ë¯¸ë§Œì¸ ìœ ì˜ë¯¸í•œ ê²°ê³¼ë§Œ í•„í„°ë§í•˜ê³ , p-valueê°€ ë‚®ì€ ìˆœìœ¼ë¡œ ì •ë ¬í•©ë‹ˆë‹¤.
significant_genes_logrank = results_df[results_df['p_value'] < 0.05].sort_values(by='p_value')

# ìµœì¢… ê²°ê³¼ ì¶œë ¥
print("\n--- ìƒì¡´ìœ¨ê³¼ ìœ ì˜ë¯¸í•œ ê´€ê³„ê°€ ìžˆëŠ” ìœ ì „ìž ëª©ë¡ (ìƒìœ„ 20ê°œ) ---")
print(significant_genes_logrank.head(20))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# \--- 1. ë°ì´í„° ì¤€ë¹„ ---



selected_genes = significant_genes_logrank['gene'].tolist()

# ì„ íƒëœ 120ê°œ ìœ ì „ìžë¡œ íŠ¹ì„±(X) ë°ì´í„°ì…‹ì„ êµ¬ì„±í•©ë‹ˆë‹¤.

X = merged_df[selected_genes]

# ì˜ˆì¸¡í•  ëŒ€ìƒ(y)ì¸ ìƒì¡´ ìƒíƒœ('Status')ë¥¼ ì§€ì •í•©ë‹ˆë‹¤.

y = merged_df['Status']

# \--- 2. í›ˆë ¨ ë° í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„ë¦¬ ---

# ë°ì´í„°ë¥¼ í›ˆë ¨ìš©(80%)ê³¼ í…ŒìŠ¤íŠ¸ìš©(20%)ìœ¼ë¡œ ë¶„ë¦¬í•©ë‹ˆë‹¤.

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2025, stratify=y
)

# \--- 3. ëª¨ë¸ í›ˆë ¨ ë° í‰ê°€ ---

# ëžœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ì„ ì´ˆê¸°í™”í•˜ê³  í›ˆë ¨í•©ë‹ˆë‹¤.

rf_model = RandomForestClassifier(n_estimators=100, random_state=2025, class_weight='balanced')
rf_model.fit(X_train, y_train)

# í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì˜ˆì¸¡ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.

y_pred = rf_model.predict(X_test)

# ìµœì¢… ëª¨ë¸ ì„±ëŠ¥ì„ ì¶œë ¥í•©ë‹ˆë‹¤.

print("--- ðŸŒ³ ìƒìœ„ 120ê°œ ìœ ì „ìž ê¸°ë°˜ ëžœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ ì„±ëŠ¥ ---")
print(f"ì •í™•ë„(Accuracy): {accuracy_score(y_test, y_pred):.3f}")
print(classification_report(y_test, y_pred))
```
